```python
from random import randint
import pandas as pd
from faker import Faker

fake = Faker()

def input_data(x):
    # pandas dataframe
    data = pd.DataFrame()
    for i in range(0, x):
        data.loc[i,'id']= randint(1, 100)
        data.loc[i,'name']= fake.name()
        data.loc[i,'address']= fake.address()
        data.loc[i,'latitude']= str(fake.latitude())
        data.loc[i,'longitude']= str(fake.longitude())
    return data

# Assign the returned DataFrame to a variable
df = input_data(1000)

# Now call to_csv on the DataFrame variable
df.to_csv('data.csv', index=False) # index=False prevents saving the index
```
